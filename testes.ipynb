```python
import pandas as pd
import streamlit as st
import plotly.express as px
```

```python
df = pd.read_csv('dados\\focos_diario_br_20251125.csv')
```

```python
df_amapa = df[df['estado'] == 'AMAPÁ']
df_amapa.head()
```

```python
fig = px.scatter_mapbox(
    df_amapa,
    lat='lat', lon='lon',
    size='frp', color='municipio',
    hover_name='municipio', hover_data=['frp', 'data_hora_gmt'],
    mapbox_style='open-street-map',
    zoom=6, center={'lat':1.5, 'lon':-52.0},
    title=f'Focos de incêndio no Amapá)'
)


fig.show()
```
